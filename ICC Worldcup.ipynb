## Top Batsmen with There Sixes

```python
six_hits = df.sort_values(by='sixes', ascending=False).head(6)
# Extract batsmen names and sixes
batsmen_names = six_hits['fullName']
sixes = six_hits['sixes']
# Create a bar plot
plt.figure(figsize=(10,4))
plt.bar(batsmen_names, sixes,color='#06C2AC')
plt.xlabel('Batsmen',fontsize=15)
plt.ylabel('Number of Sixes',fontsize=15)
plt.title('Top Batsmen with Sixes in ICC World Cup')
# Display the plot
plt.show()
```

##### The code snippet identifies the top 6 batsmen with the highest number of sixes in the dataset. It sorts the DataFrame (df) by the 'sixes' column in descending order and selects the top 6 records. The batsmen names and corresponding sixes are then extracted. Subsequently, a horizontal bar plot is created using Matplotlib, where each bar represents a batsman, and the length of the bar corresponds to the number of sixes hit. The plot is displayed with a teal color ('#06C2AC'), labeled axes, and a title indicating it showcases the top batsmen with sixes in the ICC World Cup.

##### The code creates a bar plot showing the number of balls faced by the top 10 batsmen in the ICC World Cup dataset. It extracts relevant data, sorts the DataFrame, and utilizes Matplotlib to display a red-colored bar chart with labeled axes and a title.

```python
runs_counts = df['runs'].value_counts().head(6)
run = runs_counts.tolist()
Batsman = df.groupby('fullName')['runs'].sum().sort_values(ascending=False).head(6).index.tolist()

plt.figure(figsize=(4, 8))


plt.pie(run, labels=Batsman, autopct='%1.1f%%', shadow=True, textprops={'fontsize': 13}, startangle=140)
plt.title("Distribution of Runs Scored by Batsmen",fontweight='bold')
plt.show()
```

##### The code generates a pie chart to depict the percentage distribution of runs scored by the top 6 batsmen in the DataFrame. It calculates the run counts, extracts relevant data, and creates a labeled pie chart with a title, highlighting each batsman's contribution to the total runs.

```python
fig = px.scatter_matrix(df, dimensions=["strikeRate"], color="fullName",
                        title=" Strike Rate by Batsmen",
                        labels={"strikeRate", "fullName"},
                        template='plotly_dark')

fig.show()
```

##### The code utilizes Plotly Express to create a scatter plot matrix, visualizing the relationship between the strike rate and batsmen's names. It employs a dark-themed template and colors the points based on batsmen's names for better clarity

```python
plt.figure(figsize=(9, 7))

sns.set_style("darkgrid")
sns.stripplot(x="fullName", y="minutes", data=df.head(6))

plt.title("Strip Plot of Minutes by Batsman (Top 5)")
plt.xlabel("Batsman")
plt.ylabel("Minutes")

plt.show()
```

##### The code creates a Seaborn strip plot to visualize the distribution of minutes played by the top 6 batsmen. It employs a dark grid background, and each point on the plot represents a batsman's playing time. The plot is titled and labeled for clarity.
